## 問題リスト

### 得点データの解析

- 講義資料で扱ったデータ分析を他のリーグや他のチームに適用し，結果を考察せよ．
- リーグを１つ選んで，１シーズンの総得点と総失点の散布図を作成し，得点と失点の関係を考察せよ．
- リーグを１つ選んで，全チームの1シーズンの累積勝ち点の変動を可視化せよ（横軸に節，縦軸に累積勝ち点を取ったグラフを作成する）．作成した図から分かることを自由に考察せよ．
- 全てのリーグについて，ホームチームとアウェイチームの得点分布を描画し，さらに実データから求めた平均値をパラメータとするポアソン分布を同じグラフに描画せよ．作成したグラフから分かることを自由に考察せよ．
-  $ \chi^2 $ 検定を用いて，サッカーの得点分布がポアソン分布に従うかどうか検証せよ．

### イベントデータの解析

- 講義資料で扱ったデータ分析を他のリーグや他のチームに適用し，結果を考察せよ．
- イベントログとイベントタグから好きな条件でヒートマップを描画し，特徴を分析せよ．
- 好きな条件で1シーズンの選手の個人成績を集計し，自由に考察せよ．
- パス数行列を作成してチームの特徴を分析せよ．
- ゴール期待値（xG）について調べ，簡単なxGモデルを作成せよ．
  - 例：ゴールからシュート位置までの距離や角度によってシュート成功率がどのように変化するか？

### トラッキングデータの解析

- 講義資料で扱ったデータ分析を他の試合に適用し，結果を考察せよ．
- 重心の $ x $ 座標と標準偏差の関係を可視化し，結果を考察せよ．
- 速度から速さ（速度の大きさ）を求め，選手ごとに速さのヒストグラムを求めよ．
- 各選手の走行距離を計算せよ
  - ヒント：細かい時間間隔での移動距離を全て足し合わせれば良い．
- 各選手のスプリント回数を計算せよ
  - ヒント：Jリーグでは，時速25km以上で1秒以上走った回数をスプリント回数と定義している．
- トラッキングデータのアニメーションを表示するプログラムに自由に機能を追加せよ．

### 自由課題

- 授業で扱ったデータを自由に分析し，結果を考察せよ．
- ウェブ上に公開されているオープンデータや自分で収集したデータを自由に分析し，結果を報告せよ．
  - ただし，スポーツデータに限定する．
